<h1>Варіант 7</h1>

\begin{cases}
3.2x_{1} + 8.8x_{2} & = 89 \\
-14.2x_{1} + 10.8x_{2} & = 125 \\
20.3x_{1} - 7.2x_{2} & = 142
\end{cases}

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Функції для ліній
def line1(x):
    return (89 - 3.2*x) / 8.8

def line2(x):
    return (125 + 14.2*x) / 10.8

def line3(x):
    return (142 - 20.3*x) / -7.2

# Значення x для побудови графіків
x_values = np.linspace(-20, 20, 400)

# Побудова графіків
plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.plot(x_values, line1(x_values), color='red', linestyle='--')
plt.title('3.2x1 + 8.8x2 = 89')
plt.xlabel('x1')
plt.ylabel('x2')

plt.subplot(132)
plt.plot(x_values, line2(x_values), color='green', linestyle=':')
plt.title('-14.2x1 + 10.8x2 = 125')
plt.xlabel('x1')
plt.ylabel('x2')

plt.subplot(133)
plt.plot(x_values, line3(x_values), color='blue', linestyle='-')
plt.title('20.3x1 - 7.2x2 = 142')
plt.xlabel('x1')
plt.ylabel('x2')

plt.tight_layout()
plt.show()
```

```python
# Функції для ліній
def line1(x):
    return (89 - 3.2*x) / 8.8

def line2(x):
    return (125 + 14.2*x) / 10.8

def line3(x):
    return (142 - 20.3*x) / -7.2

# Значення x для побудови графіків
x_values = np.linspace(-50, 50, 400)

# Побудова графіків
plt.figure(figsize=(8, 16), dpi=100)
plt.grid(color='gray', linestyle='--')

plt.plot(x_values, line1(x_values), color='purple', linestyle='-', label='3.2x1 + 8.8x2 = 89')
plt.plot(x_values, line2(x_values), color='orange', linestyle='--', label='-14.2x1 + 10.8x2 = 125')
plt.plot(x_values, line3(x_values), color='green', linestyle=':', label='20.3x1 - 7.2x2 = 142')

# Знаходження точок перетину
A = np.array([[3.2, 8.8], [-14.2, 10.8]])
B = np.array([89, 125])
x1, x2 = np.linalg.solve(A, B)

C = np.array([[3.2, 8.8], [20.3, -7.2]])
D = np.array([89, 142])
x3, x4 = np.linalg.solve(C, D)

plt.scatter([x1, x2, x3], [line1(x1), line2(x2), line3(x3)], color='red')
plt.annotate(f'({x1:.2f}, {line1(x1):.2f})', (x1, line1(x1)), textcoords="offset points", xytext=(-15,-10), ha='center')
plt.annotate(f'({x2:.2f}, {line2(x2):.2f})', (x2, line2(x2)), textcoords="offset points", xytext=(15,10), ha='center')
plt.annotate(f'({x3:.2f}, {line3(x3):.2f})', (x3, line3(x3)), textcoords="offset points", xytext=(15,10), ha='center')

# Заповнення області перетину
x_fill = np.linspace(max(min(x1, x2, x3), -50), min(max(x1, x2, x3), 50), 400)
y_fill1 = np.maximum(line1(x_fill), line2(x_fill))
y_fill2 = np.minimum(line2(x_fill), line3(x_fill))
plt.fill_between(x_fill, y_fill1, y_fill2, color='gray', alpha=0.5)

plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Графік системи рівнянь')
plt.legend()


plt.show()
```
